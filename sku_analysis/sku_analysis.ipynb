```python
# read csv aggregated_offtake_analysis.csv
import pandas as pd

df = pd.read_csv("../aggregated_offtake_data.csv")

# print the first 5 rows
print(df.head())

# print the column names
print(df.columns)

# print the shape of the dataframe
print(df.shape)
```

```python
# add column for the month of the bill date
df['date_datetime'] = pd.to_datetime(df['Bill Date'])
df["Month"] = df["date_datetime"].dt.month

# add column for the year of the bill date
df["Year"] = df["date_datetime"].dt.year
```

```python
df.head()
```

```python
# add column for the week of the bill date
df["Week"] = df["date_datetime"].dt.isocalendar().week
```

```python
df.head()
```

```python
# aggregate skus by month
df_monthly = df.groupby(["Month", "SKU Code"]).agg({"Qty": "sum"}).reset_index()
df_monthly.head()
# aggregate skus by week
df_weekly = df.groupby(["Week", "SKU Code"]).agg({"Qty": "sum"}).reset_index()
df_weekly.head()
```

```python
df_monthly.head()
df_monthly.shape
```

```python
sku_month_count = df_monthly[df_monthly['Qty'] > 0].groupby('SKU Code')['Month'].nunique().reset_index()
sku_month_count.columns = ['SKU Code', 'months_with_sales']

# Categorize SKUs
def categorize_sku(months_sold):
    if months_sold == 9:
        return 'Consistent (9 months)'
    elif months_sold == 8:
        return 'Frequent (8 months)'
    elif months_sold == 7:
        return 'Regular (7 months)'
    elif months_sold >= 6:
        return 'Moderate (6+ months)'
    else:
        return f'Sporadic ({months_sold} months)'

sku_month_count['category'] = sku_month_count['months_with_sales'].apply(categorize_sku)

# Get summary by category
category_summary = sku_month_count['category'].value_counts().sort_index()
print("SKU Distribution by Sales Frequency:")
print(category_summary)
print("\n")

# View SKUs in each category
for category in sorted(sku_month_count['category'].unique()):
    skus = sku_month_count[sku_month_count['category'] == category]['SKU Code'].tolist()
    print(f"\n{category}: {len(skus)} SKUs")
    print(f"SKUs: {skus[:10]}")  # Show first 10 SKUs
    if len(skus) > 10:
        print(f"... and {len(skus) - 10} more")

# Merge back to original dataframe if needed
df_with_category = df.merge(sku_month_count[['SKU Code', 'months_with_sales', 'category']], 
                             on='SKU Code', 
                             how='left')
```

```python
import pandas as pd

# First, standardize the data types - convert basepack code to string
df['Basepack Code'] = df['Basepack Code'].astype(str)
df['SKU Code'] = df['SKU Code'].astype(str)

# Now check for one-to-one mapping
sku_basepack_mapping = df[['SKU Code', 'Product Name', 'Basepack Code', 'Basepack Desc']].drop_duplicates()

# Count how many basepacks each SKU maps to
sku_to_basepack = sku_basepack_mapping.groupby('SKU Code')['Basepack Code'].nunique().reset_index()
sku_to_basepack.columns = ['SKU Code', 'basepack_count']

# Count how many SKUs each basepack maps to
basepack_to_sku = sku_basepack_mapping.groupby('Basepack Code')['SKU Code'].nunique().reset_index()
basepack_to_sku.columns = ['Basepack Code', 'sku_count']

# Identify issues
print("=" * 80)
print("MAPPING ANALYSIS (After data type standardization)")
print("=" * 80)

# Check if perfect one-to-one mapping exists
one_to_one = (sku_to_basepack['basepack_count'] == 1).all() and (basepack_to_sku['sku_count'] == 1).all()

if one_to_one:
    print("\n✓ Perfect one-to-one mapping exists!")
else:
    print("\n✗ One-to-one mapping DOES NOT exist. Issues found:\n")
    
    # Issue 1: One SKU maps to multiple basepacks
    multi_basepack_skus = sku_to_basepack[sku_to_basepack['basepack_count'] > 1]
    if not multi_basepack_skus.empty:
        print(f"\n1. SKUs mapping to MULTIPLE basepacks: {len(multi_basepack_skus)}")
        print("-" * 80)
        for _, row in multi_basepack_skus.head(20).iterrows():  # Show first 20
            sku = row['SKU Code']
            details = sku_basepack_mapping[sku_basepack_mapping['SKU Code'] == sku]
            product_name = details['Product Name'].iloc[0]
            print(f"\n   SKU Code: {sku}")
            print(f"   Product Name: {product_name}")
            print(f"   Maps to {len(details)} Basepack(s):")
            for _, detail in details.iterrows():
                print(f"      - Basepack Code: {detail['Basepack Code']}, Desc: {detail['Basepack Desc']}")
        if len(multi_basepack_skus) > 20:
            print(f"\n   ... and {len(multi_basepack_skus) - 20} more SKUs with multiple basepacks")
    
    # Issue 2: One basepack maps to multiple SKUs
    multi_sku_basepacks = basepack_to_sku[basepack_to_sku['sku_count'] > 1]
    if not multi_sku_basepacks.empty:
        print(f"\n2. Basepacks mapping to MULTIPLE SKUs: {len(multi_sku_basepacks)}")
        print("-" * 80)
        for _, row in multi_sku_basepacks.head(20).iterrows():  # Show first 20
            basepack = row['Basepack Code']
            details = sku_basepack_mapping[sku_basepack_mapping['Basepack Code'] == basepack]
            basepack_desc = details['Basepack Desc'].iloc[0]
            print(f"\n   Basepack Code: {basepack}")
            print(f"   Basepack Desc: {basepack_desc}")
            print(f"   Maps to {len(details)} SKU(s):")
            for _, detail in details.iterrows():
                print(f"      - SKU Code: {detail['SKU Code']}, Product Name: {detail['Product Name']}")
        if len(multi_sku_basepacks) > 20:
            print(f"\n   ... and {len(multi_sku_basepacks) - 20} more Basepacks with multiple SKUs")

# Summary statistics
print("\n" + "=" * 80)
print("SUMMARY")
print("=" * 80)
print(f"Total unique SKU Codes: {len(sku_to_basepack)}")
print(f"Total unique Basepack Codes: {len(basepack_to_sku)}")
print(f"SKU Codes with multiple basepacks: {len(multi_basepack_skus) if not multi_basepack_skus.empty else 0}")
print(f"Basepack Codes with multiple SKUs: {len(multi_sku_basepacks) if not multi_sku_basepacks.empty else 0}")

# Export issues to CSV if needed
if not one_to_one:
    if not multi_basepack_skus.empty:
        problematic_skus = multi_basepack_skus['SKU Code'].tolist()
        sku_issues_detail = sku_basepack_mapping[sku_basepack_mapping['SKU Code'].isin(problematic_skus)]
        sku_issues_detail = sku_issues_detail.sort_values('SKU Code')
        sku_issues_detail.to_csv('sku_multiple_basepacks.csv', index=False)
        print("\n✓ Exported: 'sku_multiple_basepacks.csv'")
    
    if not multi_sku_basepacks.empty:
        problematic_basepacks = multi_sku_basepacks['Basepack Code'].tolist()
        basepack_issues_detail = sku_basepack_mapping[sku_basepack_mapping['Basepack Code'].isin(problematic_basepacks)]
        basepack_issues_detail = basepack_issues_detail.sort_values('Basepack Code')
        basepack_issues_detail.to_csv('basepack_multiple_skus.csv', index=False)
        print("✓ Exported: 'basepack_multiple_skus.csv'")
```
